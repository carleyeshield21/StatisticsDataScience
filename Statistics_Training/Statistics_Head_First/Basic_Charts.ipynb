```python
import matplotlib.pyplot as plt

# Data
continent = ['North America','South America','Europe','Asia','Oceania','Africa','Antarctica']
sales = [1500,500,1500,2000,1000,500,1]

# Create bar chart
plt.barh(continent,sales)

# Add labels and title
plt.xlabel('sales')
plt.ylabel('continent')
plt.title('Bar Chart Example')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
score_ranges = ['0–199', '200–399', '400–599', '600–799', '800–999']
frequency = [5, 29, 56, 17, 3]

# Create bar plot
plt.bar(score_ranges, frequency, color='skyblue')

# Add labels and title
plt.xlabel('Score Range')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Scores')

# Show plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
hours = [1, 2, 2, 5, 14]
frequency = [4300, 6900, 4900, 2000, 2100]

# Create histogram with custom bins
plt.hist(hours, bins=[0, 3, 6, 9, 12, 15], weights=frequency, edgecolor='black')

# Add labels and title
plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.title('Histogram of Hours')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
data = [1] * 43 + [2] * 34 + [2] * 25 + [5] * 4 + [14] * 2

# Create histogram with different frequency densities
plt.hist(data, bins=[0.5, 1.5, 2.5, 5.5, 14.5], density=True, edgecolor='black',color='darkcyan')

# Add labels and title
plt.xlabel('Intervals')
plt.ylabel('Frequency Density')
plt.title('Histogram with Different Frequency Densities')

# Show plot
plt.show()
print(data)
```
