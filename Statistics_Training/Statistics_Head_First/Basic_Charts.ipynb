```python
import matplotlib.pyplot as plt

# Data
frequency_densities = [1, 2, 2, 5, 14]
frequencies = [4300, 6900, 4900, 2000, 2100]

# Calculate bin edges based on frequency densities
bin_edges = [0]
for i in range(len(frequency_densities)):
    print(bin_edges)
    print(bin_edges[-1], frequency_densities[i])
    bin_edges.append(bin_edges[-1] + frequency_densities[i])
    print('======')

# Create histogram
for i in range(len(frequency_densities)):
    plt.bar(bin_edges[i], frequencies[i], width=frequency_densities[i], align='edge', label=f'Width: {frequency_densities[i]}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Different Frequencies, Densities, and Widths')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
import statistics
salary = [10000,10000,10000,10000,10000,90000,90000,160000,80000,30000]

print('At original salary')
mean = statistics.mean(salary)
median = statistics.median(salary)
mode = statistics.mode(salary)
print(f'mean = {mean}\nmedian = {median}\nmode = {mode}')

print('At 2000 salary increase')
new_salary_2k = [2000+sal for sal in salary]
mean2k = statistics.mean(new_salary_2k)
median2k = statistics.median(new_salary_2k)
mode2k = statistics.mode(new_salary_2k)
print(f'mean2k = {mean2k}\nmedian2k = {median2k}\nmode2k = {mode2k}')

print('At 10% increase')
salary_10 = [sal+(sal*.1) for sal in salary]
mean10 = statistics.mean(salary_10)
median10 = statistics.median(salary_10)
mode10 = statistics.mode(salary_10)
print(f'mean10 = {mean10}\nmedian10 = {median10}\nmode10 = {mode10}')

def salary_adjustment(sweldo, salary):
    import statistics
    stdev = statistics.pstdev(salary)
    stdev_new = statistics.pstdev(sweldo)
    percent_err = ((stdev_new - stdev)/stdev)*100
    return print(f'The standard deviation will increase by {percent_err}')

print(salary_adjustment(new_salary_2k,salary))
print(salary_adjustment(salary_10,salary))
```

```python
import statistics
import numpy as np
import matplotlib.pyplot as plt

ppg = [3]*2 + [6]*1 + [7]*2 + [10]*3 + [11]*1 + [13]*1 + [30]*1
print(ppg)
range = max(ppg) - min(ppg)
print(f'range = {range}')

q1 = np.percentile(ppg,25)
q2 = np.percentile(ppg,50)
q3 = np.percentile(ppg,75)

print(f'q1 = {q1}\nq3 = {q3}')
print(f'q2 = {q2}')
print(f'IQR = {q3-q1}')

plt.boxplot(ppg)
plt.show()
```

```python
import statistics
p1_score = [7]*1 + [9]*2 + [10]*4 + [11]*2 + [13]*1
p2_score = [7]*1 + [8]*1 + [9]*2 + [10]*2 + [11]*2 + [12]*1 + [13]*1
p3_score = [3]*2 + [6]*1 + [7]*2 + [10]*3 + [11]*1 + [13]*1 + [30]*1

def game(scores):
    import statistics

    mean = statistics.mean(scores)
    variance = statistics.pvariance(scores)
    stdev = statistics.pstdev(scores)
    return print(f'mean = {mean}\nvariance = {variance}\nstdev = {stdev}')

print(game(p1_score))
print(game(p2_score))
print(game(p3_score))
```
