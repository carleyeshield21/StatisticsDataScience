```python
import matplotlib.pyplot as plt

# Data
continent = ['North America','South America','Europe','Asia','Oceania','Africa','Antarctica']
sales = [1500,500,1500,2000,1000,500,1]

# Create bar chart
plt.barh(continent,sales)

# Add labels and title
plt.xlabel('sales')
plt.ylabel('continent')
plt.title('Bar Chart Example')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
score_ranges = ['0–199', '200–399', '400–599', '600–799', '800–999']
frequency = [5, 29, 56, 17, 3]

# Create bar plot
plt.bar(score_ranges, frequency, color='skyblue')

# Add labels and title
plt.xlabel('Score Range')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Scores')

# Show plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [10, 20, 15, 25]
widths = [0.5, 0.8, 0.6, 0.4]  # Specify the width of each bar

# Create bar chart with different widths
plt.bar(categories, values, width=widths)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart with Different Widths')

# Show plot
plt.show()
```
