```python
sorted_roa = sorted(roa.unique())
len(sorted_roa)

for index in range(1, len(sorted_roa)-1):
    print(f'{sorted_roa[index]}:{sorted_roa[index+1]}')
    if (sorted_roa[index+1] - sorted_roa[index]) != 1:
        print(sorted_roa[index] + 1)
```

```python
# Using list comprehension
missing_nums = [sorted_roa[index] + 1 for index in range(1, len(sorted_roa)-1) if (sorted_roa[index+1] - sorted_roa[index]) != 1]
print(missing_nums)
```

### Creating a dummies column

```python
reasons_column = pandas.get_dummies(data['Reason for Absence']).astype(int)
# reasons_column
```

### Checking the sum of each row if equals 1

```python
# Add a new column named 'check' first to the reasons_column dataframe, axis=1 for rows
reasons_column['check'] = reasons_column.sum(axis=1)
```

```python
reasons_column
```

### To check if the sum of the check column is equal to the total number of rows in the table

```python
reasons_column['check'].sum(axis=0)
```

### To check if there is only one unique value

```python
reasons_column['check'].unique()
```

### After verifying the contents, we can now remove the check column

```python
reasons_column = reasons_column.drop(['check'], axis=1)
reasons_column
```
