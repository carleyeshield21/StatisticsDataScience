```python
import pandas
```

```python
fos = pandas.read_csv('field_of_study.csv')
fos_copy = fos.copy()
```

```python
fos_copy
```

```python
import matplotlib.pyplot as plt

categories = fos_copy['MAJOR FIELD OF STUDY']
values1 = fos_copy.MALES
values2 = fos_copy.FEMALES

# Set up the figure and axis
fig, ax = plt.subplots()

# Plotting the bar graph
ax.bar(categories, values1, label='Bar 1', color='blue')
# ax.bar(categories, values2, label='Bar 2', color='orange', bottom=values1)
ax.bar(categories, values2, label='Bar 2', color='orange')

# Adding labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Bar Graph with Two Bars Example')
plt.xticks(rotation=45, ha='right')


# Adding legend
ax.legend()

# Show the plot
plt.show()
```
