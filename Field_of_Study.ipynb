```python
import pandas
```

```python
fos = pandas.read_csv('field_of_study.csv')
fos_copy = fos.copy()
```

```python
fos_copy
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
categories = fos_copy['MAJOR FIELD OF STUDY']
values1 = fos_copy.MALES
values2 = fos_copy.FEMALES

# Set up the figure and axis
fig, ax = plt.subplots()

# Calculate the positions for the bars
bar_width = 0.35
bar_positions1 = np.arange(len(categories))
bar_positions2 = bar_positions1 + bar_width

# Plotting the first bar graph
bar1=ax.bar(bar_positions1, values1, width=bar_width, label='Graph 1', color='blue')

# Plotting the second bar graph
bar2=ax.bar(bar_positions2, values2, width=bar_width, label='Graph 2', color='orange')

# Adding labels and title
ax.set_xlabel('Fields of Study')
ax.set_ylabel('Number of students')
ax.set_title('Major Fields of Study between Males and Females')
plt.xticks(rotation=45, ha='right')

# Adding legend
# ax.legend()
ax.legend([bar1, bar2], ['Males', 'Females'], loc='upper right', title='Graph Legends')


# Customize x-axis tick labels
ax.set_xticks(bar_positions1 + bar_width / 2)
ax.set_xticklabels(categories)

# Show the plot
plt.show()
```
