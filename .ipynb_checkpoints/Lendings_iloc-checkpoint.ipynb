```python
import pandas
import numpy
```

```python
data = pandas.read_csv('Lending-company.csv',index_col='StringID')
```

```python
lending_data = data.copy()
lending_data.head()
```

```python
lending_data.TotalDaysYr
```

```python
# Using the iloc method to access a specific row using the index
sample1 = lending_data.iloc[3]
sample1
```

```python
# To access a specific cell item in the iloc row
lending_data.iloc[3,11]
```

```python
sample2 = lending_data.iloc[3,:]
sample2
```

```python
# 
lending_data.iloc[:,[3,1]]
```

```python
# This will output the row on index3 with all the columns, if a column specifier is not given,
# the default will be all columns until it is specified, exceeding the number of columns will output all the columns
lending_data.iloc[[3],:]
```

```python
# This will output the specified row in the indices with all the columns, we can add more to the row specifier if we want
lending_data.iloc[[8,0,56],:]
```

```python
# This will output the row on index21 with 4 columns in the column specifier
lending_data.iloc[[21],:4]
```
